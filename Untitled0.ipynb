```python
#bài14
import random
#a. Tạo mảng ngẫu nhiên
def tao_mang(m, n):
    return [[random.randint(1, 100) for _ in range(n)] for _ in range(m)]
#b. Xuất dòng k
def xuat_dong(a, k):
    print(a[k])
#c. Xuất cột k
def xuat_cot(a, k):
    for row in a:
        print(row[k], end=" ")
#d. Dòng có tổng lớn nhất
def dong_max(a):
    sums = [sum(row) for row in a]
    return sums.index(max(sums))
#e. Cột có tích nhỏ nhất
def cot_min(a):
    m, n = len(a), len(a[0])
    min_tich = float('inf')
    index = 0

    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]

        if tich < min_tich:
            min_tich = tich
            index = j

    return index
#f. Dòng chẵn, cột lẻ
def chan_le(a):
    for i in range(0, len(a), 2):
        for j in range(1, len(a[0]), 2):
            print(a[i][j], end=" ")
#g. TBC số chẵn dòng lẻ
def tbc_chan_dong_le(a):
    s = 0
    dem = 0

    for i in range(1, len(a), 2):
        for x in a[i]:
            if x % 2 == 0:
                s += x
                dem += 1

    return s/dem if dem != 0 else 0
#h. TBC phần tử biên
def tbc_bien(a):
    m, n = len(a), len(a[0])
    s = 0
    dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                s += a[i][j]
                dem += 1

    return s/dem
#i. TBC phần tử không biên
def tbc_trong(a):
    m, n = len(a), len(a[0])
    s = 0
    dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            s += a[i][j]
            dem += 1

    return s/dem if dem != 0 else 0
```
